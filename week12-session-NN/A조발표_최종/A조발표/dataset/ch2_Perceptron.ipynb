 NAND GATE OR GATE AND GATE  XOR   .

```python
import numpy as np

#y = w1*x1 + w2*x2 + bias
#y  0  1  0  0 . 
def operator(x1, x2, w1, w2, b):

    x = np.array([x1, x2])
    w = np.array([w1, w2])
    y = np.sum(w*x) + b

    return 1 if y > 0 else 0

#AND 
def AND(x1, x2):

    return operator(x1, x2, 20, 20, -30)

#OR 
def OR(x1, x2):

    return operator(x1, x2, 20, 20, -10)

#NAND 
def NAND(x1, x2):

    return operator(x1, x2, -10, -10, 15)

#XOR 
def XOR(x1, x2):

    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    return AND(s1, s2)

xs = [(0, 0), (0, 1), (1, 0), (1, 1)]

#AND     
#OUTPUT : 0 0 0 1
for x1, x2 in xs:
    print(AND(x1, x2))
```

```python
#OR     
#OUTPUT: 0 1 1 1
for x1, x2 in xs:
    print(OR(x1, x2))
```

```python
#NAND     
#OUTPUT : 1 1 1 0
for x1, x2 in xs:
    print(NAND(x1, x2))
```

```python
#XOR     
#OUTPUT : 0 1 1 0
for x1, x2 in xs:
    print(XOR(x1, x2))
```

XNOR   !

```python
#  .
def NOR(x1, x2):
    return operator(x1, x2, -20, -20, 10)

def XNOR(x1, x2):
    s1 = AND(x1, x2)
    s2 = NOR(x1, x2)
    return OR(s1, s2)

for x1, x2 in xs:
    print(XNOR(x1, x2))
```
