기존의 선형함수로 XOR 데이터를 설명하는 것은 불가능 하다.
따라서 NAND GATE와 OR GATE AND GATE를 이용하여 XOR 문제를 풀어야 한다.

```python
import numpy as np

#y = w1*x1 + w2*x2 + bias
#y 값이 0보다 크면 1을 반환하고 0보다 작으면 0을 반환한다. 
def operator(x1, x2, w1, w2, b):

    x = np.array([x1, x2])
    w = np.array([w1, w2])
    y = np.sum(w*x) + b

    return 1 if y > 0 else 0

#AND 게이트
def AND(x1, x2):

    return operator(x1, x2, 20, 20, -30)

#OR 게이트
def OR(x1, x2):

    return operator(x1, x2, 20, 20, -10)

#NAND 게이트
def NAND(x1, x2):

    return operator(x1, x2, -10, -10, 15)

#XOR 게이트
def XOR(x1, x2):

    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    return AND(s1, s2)

xs = [(0, 0), (0, 1), (1, 0), (1, 1)]

#AND 게이트를 실행시켰을 때의 결과 값
#OUTPUT : 0 0 0 1
for x1, x2 in xs:
    print(AND(x1, x2))
```

```python
#OR 게이트 실행시켰을 때의 결과 값
#OUTPUT: 0 1 1 1
for x1, x2 in xs:
    print(OR(x1, x2))
```

```python
#NAND 게이트 실행시켰을 때의 결과 값
#OUTPUT : 1 1 1 0
for x1, x2 in xs:
    print(NAND(x1, x2))
```

```python
#XOR 게이트 실행시켰을 때의 결과 값
#OUTPUT : 0 1 1 0
for x1, x2 in xs:
    print(XOR(x1, x2))
```

XNOR을 퍼셉트론을 이용하여 풀어보자!

```python
#코드를 작성해 봅시다.
def NOR(x1, x2):
    return operator(x1, x2, -20, -20, 10)

def XNOR(x1, x2):
    s1 = AND(x1, x2)
    s2 = NOR(x1, x2)
    return OR(s1, s2)

for x1, x2 in xs:
    print(XNOR(x1, x2))
```
