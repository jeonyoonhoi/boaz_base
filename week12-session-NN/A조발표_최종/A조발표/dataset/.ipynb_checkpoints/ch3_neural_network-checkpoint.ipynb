#  활성화 함수 구현
##### 활성화 함수 : 입력 신호의 총합이 노드의 활성화를 일으키는지를 정하는 역할

```python
# ① 시그모이드 함수
```

```python
# 시그모이드 함수 정의
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# 시그모이드 함수 그래프 확인
X = np.arange(-5.0, 5.0, 0.1)
Y = sigmoid(X)

plt.plot(X, Y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# ② 계단 함수
```

```python
# 계단 함수 정의
def step_function(x):
    return np.array(x > 0, dtype=np.int)
```

```python
# 계단 함수 그래프 확인
X = np.arange(-5.0, 5.0, 0.1)
Y = step_function(X)

plt.plot(X, Y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# 시그모이드, 계단 함수 비교
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x)
y2 = step_function(x)

plt.plot(x, y1)
plt.plot(x, y2, 'k--')
plt.ylim(-0.1, 1.1)
plt.show()

### 계단 함수가 0과 1 값만을 반환하는 반면에, 
### 시그모이드 함수는 실수(0.731, 0.880, .......)를 반환해준다.
### but, 둘의 공통점은 
### ① 입력이 작을 때 출력이 0에 가깝고, 입력이 커지면 1에 가까워지는 구조!
### ② 출력값이 0과 1 사이
```

```python
# ③ ReLU 함수
```

```python
# ReLU 함수 정의
def relu(x):
    return np.maximum(0, x)
```

```python
# ReLU 함수 그래프 확인
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-1.0, 5.5)
plt.show()

### 입력이 0을 넘으면 그 입력을 그대로 출력하고, 0 이하이면 0을 출력하는 함수.
### 두 수 중 큰값을 반환하는 np.maximum 함수를 이용해 구현.
```

# 3층 신경망 구현

```python
Image("2.jpg")
```

```python
# 신경망 초기화 함수
def init_network():
    network = {}
    ### 초기값을 포함해서 신경망 초기화
    network['W1'] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
    network['b1'] = np.array([0.1, 0.2, 0.3])
    network['W2'] = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
    network['b2'] = np.array([0.1, 0.2])
    network['W3'] = np.array([[0.1, 0.3], [0.2, 0.4]])
    network['b3'] = np.array([0.1, 0.2])
    return network

# 최종값 출력 함수
def forward(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']
    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = a3
    return y

network = init_network()
x = np.array([1.0, 0.5])
y = forward(network, x)

print(y)
```

# 출력층 설계 - 소프트맥스 함수 구현

```python
### 기계학습 문제는 분류classification와 회귀regression로 나뉜다.
### 분류 : 데이터가 어느 클래스에 속하는가 하는 문제. ex.사진속 인물의 성별
### 회귀 : 입력 데이터에서 (연속적인) 수치를 예측하는 문제. ex. 사진속 인물의 몸무게

### 기계학습을 통해 분류를 할 경우 → 소프트맥스 함수를 출력함수로 사용
### 기계학습을 통해 회귀를 할 경우 → 항등 함수를 출력함수로 사용

### ※항등함수의 경우, 말 그대로 출력값을 그대로 다시 출력해주는 함수이기 때문에 여기서는 따로 정리 X
```

```python
Image("3.jpg")
```

```python
# 소프트맥스 함수 정의
def softmax(a):
    ### 오버플로 대책
    #c = np.max(a)
    #exp_a = np.exp(a - c)
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
# 오버플로우 문제 확인
print(softmax(np.array([1010, 1000, 990])))

#### [ nan  nan  nan] → 오버플로 발생
#### [  9.99954600e-01   4.53978686e-05   2.06106005e-09] → 오버플로 막은 후 결과
```

```python
# 소프트맥스 함수의 특징 : 출력값을 다 더하면 1
a = np.array([0.3, 2.9, 4.0])
y = softmax(a)
print(y)
print(np.sum(y))

### 출력값을 확률로 해석 가능 (소프트맥스 함수의 형태를 살펴보기)
### 입력값이 특정 클래스로 분류될 확률을 의미
### 따라서 출력층의 노드 수를 결정할 때는 특정 문제를 몇개의 클래스로 분류할 지를 고려해야 함
### 예를 들어, 숫자 손글씨 이미지를 0~9로 분류하는 문제에서는 출력층을 10개로 설정함
```

# MNIST 손글씨 데이터셋으로 실습
##### 0~9 숫자 이미지로 구성.
##### 훈련 이미지 60000장, 시험 이미지 10000장
##### (훈련 이미지로 학습하고, 시험이미지로 얼마나 잘 분류하는 지 측정)
#####  28*28 크기의 회색조 이미지이며 각 픽셀은 0~255의 값을 가짐
##### 레이블은 정답에 해당하는 숫자

```python
# MNIST 이미지 확인

import sys, os
sys.path.append(os.pardir)
from dataset.mnist import load_mnist
from PIL import Image
def img_show(img):
    pil_img = Image.fromarray(np.uint8(img))
    pil_img.show()

(x_train, t_train), (x_test, t_test) = load_mnist(flatten=True,normalize=False)

img = x_train[0]
label = t_train[0]
print(label)
print(img.shape)  ### (784,)
img = img.reshape(28, 28)  ### 원래 이미지 모양으로 변형
print(img.shape)  ### (28, 28)
img_show(img)
```

```python
# MNIST 데이터셋 읽어들이기

import pickle
from common.functions import sigmoid, softmax

### 처음 한 번은 몇 분 정도 걸림
### 이미지를 numpy 배열로 저장
### flatten : 입력 이미지를 평탄화(1차원 배열로)
### normalize : 입력 이미지를 정규화(0.0~1.0 사이의 값으로) → false로 설정하면 원래값 그대로 0~255사이값 유지
### one_hot_label : 레이블을 원-핫 인코딩 형태로 저장 → 정답을 뜻하는 원소만 1이고, 나머지는 모두 0인 배열 (더미 변수처럼)

(x_train, t_train), (x_test, t_test) = load_mnist(flatten=True, normalize=False)

# 각 데이터의 형상 출력

print(x_train.shape)  ### (60000, 784) ※ 28X28 = 784 → flatten 옵션을 통해 28X28 이미지를 1차원 배열로 바꿨으므로 784
print(t_train.shape)  ### (60000,) ※ 라벨
print(x_test.shape)  ### (10000, 784)
print(t_test.shape)  ### (10000,)
```

```python
# 미리 학습된 가중치 데이터 불러들이기

def get_data():
    (x_train, t_train), (x_test, t_test) = load_mnist(flatten=True, normalize=True, one_hot_label=False)
    return x_test, t_test

def init_network():
    with open("sample_weight.pkl", 'rb') as f : network = pickle.load(f)
    ### sample_weight.pkl → 미리 학습된 가중치 매개변수가 담긴 파일
    ### 학습 없이 바로 추론을 수행
    return network

network = init_network()
                           ### <알 수 있는 정보>
print(network['W1'].shape) ### 입력층 : 노드 784개 / 첫번째 은닉층 : 노드 50개
print(network['W2'].shape) ### 첫번째 은닉층 : 노드 50개 / 두번째 은닉층 : 노드 100개
print(network['W3'].shape) ### 두번째 은닉층 : 노드 100개 / 출력층 : 노드 10개 (분류해야할 숫자의 개수가 10개이므로)
```

```python
# 신경망 구성

def predict(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']
    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = softmax(a3)
    return y
```

```python
# 결과 샘플 확인

x, t = get_data()
network = init_network()
accuracy_cnt = 0

y_sample = predict(network, x[0])
print(y_sample)
print(np.argmax(y_sample)) ### np.argmax : 배열에서 가장 큰 값의 인덱스를 반환하는 함수 (즉, 여기서는 확률이 가장 높은 원소의 인덱스를 반환)
```

```python
# 전체 결과의 정확성 확인 

for i in range(len(x)):
    y = predict(network, x[i])
    p = np.argmax(y)
    if p == t[i] : accuracy_cnt += 1 ### 예측값과 실제값이 맞을시 카운트 1 증가

print("Accuracy:" + str(float(accuracy_cnt) / len(x)))  ### Accuracy:0.9352
```

```python
# 배치(batch)를 통해 효율적으로 계산(컴퓨터의 연산에 적합)

accuracy_cnt = 0
batch_size = 100

for i in range(0, len(x), batch_size): ### range(초기값, 끝값, 초기값에서 몇 씩 증가시킬 것인가)
    x_batch = x[i:i+batch_size]
    y_batch = predict(network, x_batch)
    p = np.argmax(y_batch, axis=1)
    accuracy_cnt += np.sum(p == t[i:i+batch_size])

print("Accuracy:" + str(float(accuracy_cnt) / len(x)))  # Accuracy:0.9352
```
